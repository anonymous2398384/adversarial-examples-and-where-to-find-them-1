# Comparison of robustness curves for l_2 and l_inf norms and models trained with different robust training methods

## Calculate robustness curves:
Estimated runtime (if no file with data is present): 21 days

```python
def load_from_json(file_name):

    if not os.path.exists("res/" + file_name + ".json"):
        return None
    else:
        with open("res/" + file_name + ".json", 'r') as fp:
            loaded_json =  json.load(fp)

            for key in loaded_json.keys():
                loaded_json[key]["x"] = np.array(loaded_json[key]["x"])
                loaded_json[key]["y"] = np.array(loaded_json[key]["y"])

                loaded_json[key]["y"][np.isnan(loaded_json[key]["x"])] = 1.0
                loaded_json[key]["x"] = np.nan_to_num(loaded_json[key]["x"], nan = np.nanmax(loaded_json[key]["x"]))

            return loaded_json
        
def save_to_json(dictionary, file_name):
        
    if not os.path.exists("res"):
        os.makedirs("res")

    with open("res/" + file_name + ".json", 'w') as fp:
        json.dump(dictionary, fp, cls = NumpyEncoder)


training_method_to_model_path = {"ST": "provable_robustness_max_linear_regions/models/plain/2019-02-24 00:50:45 dataset=mnist nn_type=cnn_lenet_small p_norm=2 lmbd=0.0 gamma_rb=0.0 gamma_db=0.0 ae_frac=0.0 lr=0.001 epoch=100.mat",
                                 "MMR+AT_l_inf": "provable_robustness_max_linear_regions/models/mmr+at/2019-02-17 01:54:16 dataset=mnist nn_type=cnn_lenet_small p_norm=inf lmbd=0.5 gamma_rb=0.2 gamma_db=0.2 ae_frac=0.5 epoch=100.mat", 
                                 "MMR+AT_l_2": "provable_robustness_max_linear_regions/models/mmr+at/2019-02-24 00:04:27 dataset=mnist nn_type=cnn_lenet_small p_norm=2 lmbd=6.0 gamma_rb=0.45 gamma_db=0.45 ae_frac=0.5 lr=5e-05 epoch=50.mat", 
                                 "KW_l_inf": "provable_robustness_max_linear_regions/models/kw/p_norm=inf dataset=mnist_model=cnn_lenet_small_method=robust_eps=0.1_checkpoint.mat", 
                                 "KW_l_2": "provable_robustness_max_linear_regions/models/kw/p_norm=2 dataset=mnist_model=cnn_lenet_small_method=robust_eps=0.3_checkpoint.mat", 
                                 "AT_l_inf": "provable_robustness_max_linear_regions/models/at/2019-02-19 01:20:16 dataset=mnist nn_type=cnn_lenet_small p_norm=inf lmbd=0.0 gamma_rb=0.0 gamma_db=0.0 ae_frac=0.5 epoch=100.mat", 
                                 "AT_l_2": "provable_robustness_max_linear_regions/models/at/2019-02-24 00:50:45 dataset=mnist nn_type=cnn_lenet_small p_norm=2 lmbd=0.0 gamma_rb=0.0 gamma_db=0.0 ae_frac=0.5 lr=0.001 epoch=100.mat",
                                 "MMRUNIV": "experiments/additional_models/mmr_univ_mnist_gammas_1.0_0.175_lambdas_3.0_12.0.mat"}

n_points = 10000

robustness_curve_data = dict()

for training_method in ["ST", "MMR+AT_l_inf", "MMR+AT_l_2", "KW_l_inf", "KW_l_2", "AT_l_inf", "AT_l_2", "MMRUNIV"]:
    
    robustness_curve_data[training_method] = load_from_json("rob_curve_data_{}_n_points={}".format(training_method, n_points))
                                           
    if not robustness_curve_data[training_method]:

        args = Namespace()

        args.dataset = "mnist"
        args.n_points = n_points
        args.model_path = training_method_to_model_path[training_method]
        args.nn_type = "cnn"
        args.norms = ["inf", "2"]
        args.save = False
        args.plot = False

        robustness_curve_data[training_method] = generate_curve_data(args)

        save_to_json(robustness_curve_data[training_method], "rob_curve_data_{}_n_points={}".format(training_method, n_points))     
```

## Plot:

```python
# name to save the plot
save_name = "fig_rc_model_comparison"

training_method_to_color = {"ST": colors["blue"],
                             "MMR+AT_l_inf": colors["red"], 
                             "MMR+AT_l_2": colors["green"],
                             "MMRUNIV": colors["yellow"],
                             "KW_l_inf": colors["red"], 
                             "KW_l_2": colors["green"], 
                             "AT_l_inf": colors["red"], 
                             "AT_l_2": colors["green"]}

training_method_to_col_title = {"ST": "Training: Standard Training",
                             "MMR+AT_l_inf": "Training: MMR+AT\nThreat Model: $\ell_\infty(\epsilon=0.1)$", 
                             "MMR+AT_l_2": "Training: MMR+AT\nThreat Model: $\ell_2(\epsilon=0.3)$",
                             "MMRUNIV": "Training: MMR UNIVERSAL\nThreat Model: all $\ell_p$ norms",
                             "KW_l_inf": "Training: KW\nThreat Model: $\ell_\infty(\epsilon=0.1)$", 
                             "KW_l_2": "Training: KW\nThreat Model: $\ell_2(\epsilon=0.3)$", 
                             "AT_l_inf": "Training: AT\nThreat Model: $\ell_\infty(\epsilon=0.1)$", 
                             "AT_l_2": "Training: AT\nThreat Model: $\ell_2(\epsilon=0.3)$"}



color_map = [colors["blue"], colors["red"], colors["green"]]

# number of model types and parameter combinations
n_cols = 3
n_rows = 2

fig, ax = plt.subplots(n_rows, n_cols, figsize = (6 * n_cols, 5 * n_rows), sharey = 'row', sharex = "row", dpi=400)

for training_method in ["ST", "AT_l_inf", "AT_l_2"]:
    ax[0][0].plot(robustness_curve_data[training_method]["inf"]["x"], robustness_curve_data[training_method]["inf"]["y"] * 100, c = training_method_to_color[training_method], label = "{}".format(training_method_to_col_title[training_method]))
    ax[1][0].plot(robustness_curve_data[training_method]["2"]["x"], robustness_curve_data[training_method]["2"]["y"] * 100, c = training_method_to_color[training_method], label = "{}".format(training_method_to_col_title[training_method]))
    
ax[0][0].legend(loc="lower right")
ax[1][0].legend(loc="lower right")
                                                        
for training_method in ["ST", "KW_l_inf", "KW_l_2"]:
    ax[0][1].plot(robustness_curve_data[training_method]["inf"]["x"], robustness_curve_data[training_method]["inf"]["y"] * 100, c = training_method_to_color[training_method], label = "{}".format(training_method_to_col_title[training_method]))
    ax[1][1].plot(robustness_curve_data[training_method]["2"]["x"], robustness_curve_data[training_method]["2"]["y"] * 100, c = training_method_to_color[training_method], label = "{}".format(training_method_to_col_title[training_method]))
    
ax[0][1].legend(loc="lower right")
ax[1][1].legend(loc="lower right")
                                                         
for training_method in ["ST", "MMR+AT_l_inf", "MMR+AT_l_2", "MMRUNIV"]:
    ax[0][2].plot(robustness_curve_data[training_method]["inf"]["x"], robustness_curve_data[training_method]["inf"]["y"] * 100, c = training_method_to_color[training_method], label = "{}".format(training_method_to_col_title[training_method]))
    ax[1][2].plot(robustness_curve_data[training_method]["2"]["x"], robustness_curve_data[training_method]["2"]["y"] * 100, c = training_method_to_color[training_method], label = "{}".format(training_method_to_col_title[training_method]))

ax[0][0].set_ylabel("test set loss in $\%$")
ax[1][0].set_ylabel("test set loss in $\%$")

ax[0][2].legend(loc="lower right")
ax[1][2].legend(loc="lower right")
ax[0][0].set_xlabel("perturbation size (measured in $\ell_{\infty}$ norm)")
ax[0][1].set_xlabel("perturbation size (measured in $\ell_{\infty}$ norm)")
ax[0][2].set_xlabel("perturbation size (measured in $\ell_{\infty}$ norm)")
ax[1][0].set_xlabel("perturbation size (measured in $\ell_{2}$ norm)")
ax[1][1].set_xlabel("perturbation size (measured in $\ell_{2}$ norm)")
ax[1][2].set_xlabel("perturbation size (measured in $\ell_{2}$ norm)")

ax[0][0].set_title("AT")
ax[0][1].set_title("KW")
ax[0][2].set_title("MMR+AT")
                                                         
fig.tight_layout()
fig.savefig('res/{}.pdf'.format(save_name))
```
